## Exercice 3 — Simulation boursière vectorisée

#### Partie 1 — Génération

```python
import numpy as np

variations = np.random.uniform(low=1-0.03, high=1+0.03, size=(1000,364))
variations = np.concatenate((np.ones((1000,1)), variations), axis=1)

actions = np.random.uniform(40, 260, (1000, 1))

print(variations.shape)
#print(variations)
#print(actions)
```

#### Partie 2 — Calculs

rendement moyen

Le **rendement moyen**, ou retour sur investissement moyen, représente le gain financier annuel réalisé sur un investissement pendant une période donnée, servant à évaluer la performance et à comparer le risque. Il existe deux méthodes principales de calcul selon le contexte :

*   **Taux de rendement moyen (TRM) ou comptable (ARR)** : Utilisé en comptabilité et budgétisation des immobilisations, il compare le bénéfice annuel moyen au coût initial. La formule est : $$\text{TRM} = \left( \frac{\text{Bénéfice annuel moyen}}{\text{Coût de l'investissement}} \right) \times 100\%$$. Cette méthode est simple mais ne tient pas compte de la valeur temporelle de l'argent ni de la volatilité.
*   **Rendement moyen annualisé (géométrique)** : Méthode privilégiée pour les investissements financiers (bourse, ETF) car elle intègre l'effet de composition. Pour un investissement unique sans flux de trésorerie, la formule est : $$\text{Rendement annualisé} = \left( \frac{\text{Valeur finale}}{\text{Valeur initiale}} \right)^{\frac{1}{\text{Années}}} - 1$$. Elle reflète la croissance réelle du capital au fil du temps, contrairement à la moyenne arithmétique qui a tendance à surestimer les résultats en cas de forte volatilité.

Les rendements historiques varient selon l'actif : les **actions** offrent en moyenne **7 à 10 %** nominal (soit **5 à 7 %** réel après inflation) sur le long terme, tandis que les **obligations** ou le **crowdfunding** doivent offrir un rendement supérieur au fonds euros pour compenser leur risque. Pour les portefeuilles avec des dépôts et retraits irréguliers, l'**IRR** (Internal Rate of Return) ou **XIRR** est la méthode la plus précise pour calculer le rendement annualisé pondéré en capital.

```python
variation_totale = variations.prod(axis=1)

#rendement moyen
avg_rendement = (variation_totale / variations[:, 0]) -1
#print(avg_rendement)

#pire action - ceci représente la variation de l'action initiale (soit 1)
worst_action = variation_totale.min()

#meilleure action
best_action = variation_totale.max()

#volatilité pour chaque action (std)
volatilité = variations.std(axis=1)
#print(volatilité)

print(worst_action)
print(best_action)
print('la meilleure action est à l\'indice:', np.where(variation_totale == best_action), '\n')
```

#### Partie 3 — Simulation cumulée

```python
variations_cumprod = variations.cumprod(axis=1)
action_everyday =  actions * variations_cumprod
print(action_everyday)
```
