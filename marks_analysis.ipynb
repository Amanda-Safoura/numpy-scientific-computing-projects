## Exercice 1 - Analyse de notes d’étudiants

```python
import numpy as np

np.random.seed(455) # fixer la seed
notes = np.random.randint(0,20, size=(84, 8))

print('Informations générales')
print(f'shape: {notes.shape}')
print(f'dtype: {notes.dtype}')

notes = notes.astype(np.int8)
print(f'\ntype après modification: {notes.dtype}')

print("nombres d'étudiants: " + str(notes.shape[0]))
print("nombres de matières: " + str(notes.shape[1]))


print('\nTableau de notes\n')
print(notes)
```

#### Partie 2 — Statistiques

```python
#Calculer :
    #moyenne globale
    #moyenne par étudiant
    #moyenne par matière
    #note maximale par matière
    #note minimale pour la 5ème matière

mean_all = notes.mean()
print("moyenne globale:", str(mean_all))

student_mean = notes.mean(axis=1)
print('\nmoyenne par étudiant: {}'.format(student_mean))

mat_mean = notes.mean(axis=0)
print('\nmoyenne par matière: {}'.format(mat_mean))

max_note_per_mat = notes.max(axis=0, keepdims=True)
print(f'\nnote maximale par matière: \n{max_note_per_mat}')

min_mat_5 = notes[4].min()
print('\nnote minimale pour la 5ème matière:', str(min_mat_5))
```

#### Partie 3 — Filtrage avec masques

```python
"""Créer :
- un masque des notes > 10
- un masque des étudiants admis (moyenne >= 10)"""


mask_great_10 = notes>10
notes_sup_10 = notes[mask_great_10]

mask_admis = notes.mean(axis=1) >= 10
print(mask_admis)
nb_admis = notes[mask_admis].shape[0]
print("On a donc : {} admis".format(nb_admis))
```

#### Partie 4 — Bonus NumPy réel

```python
#Ajouter un bonus de : +2 points à toutes les notes avec broadcasting

notes = notes + 2
notes = np.where(notes > 20, 20, notes)
print(notes)
```
