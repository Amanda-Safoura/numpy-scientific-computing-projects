## Exercice 1 - Analyse de notes d’étudiants

```python
import numpy as np

np.random.seed(455) # fixer la seed
notes = np.random.randint(0,20, size=(84, 8))

print('Informations générales')
print(f'shape: {notes.shape}')
print(f'dtype: {notes.dtype}')

notes = notes.astype(np.int8)
print(f'\ntype après modification: {notes.dtype}')

print("nombres d'étudiants: " + str(notes.shape[0]))
print("nombres de matières: " + str(notes.shape[1]))


print('\nTableau de notes\n')
print(notes)
```

#### Partie 2 — Statistiques

```python
#Calculer :
    #moyenne globale
    #moyenne par étudiant
    #moyenne par matière
    #note maximale par matière
    #note minimale pour la 5ème matière

mean_all = notes.mean()
print("moyenne globale:", str(mean_all))

student_mean = notes.mean(axis=1)
print('\nmoyenne par étudiant: {}'.format(student_mean))

mat_mean = notes.mean(axis=0)
print('\nmoyenne par matière: {}'.format(mat_mean))

max_note_per_mat = notes.max(axis=0, keepdims=True)
print(f'\nnote maximale par matière: \n{max_note_per_mat}')

min_mat_5 = notes[4].min()
print('\nnote minimale pour la 5ème matière:', str(min_mat_5))
```

#### Partie 3 — Filtrage avec masques

```python
"""Créer :
- un masque des notes > 10
- un masque des étudiants admis (moyenne >= 10)"""


mask_great_10 = notes>10
notes_sup_10 = notes[mask_great_10]

mask_admis = notes.mean(axis=1) >= 10
print(mask_admis)
nb_admis = notes[mask_admis].shape[0]
print("On a donc : {} admis".format(nb_admis))
```

#### Partie 4 — Bonus NumPy réel

```python
#Ajouter un bonus de : +2 points à toutes les notes avec broadcasting

notes = notes + 2
notes = np.where(notes > 20, 20, notes)
print(notes)
```

convertir en niveau de gris

La conversion en niveau de gris consiste à remplacer les trois valeurs de couleur (Rouge, Vert, Bleu) de chaque pixel par une seule valeur de luminosité, créant ainsi une image monochrome. Cette conversion réduit le nombre d'octets nécessaires au codage et s'oppose au noir et blanc pur (pixels soit noirs, soit blancs) en offrant des nuances intermédiaires.

### Méthodes de Conversion

*   **Algorithme de Luminance (Standard)** : Méthode la plus précise qui tient compte de la perception humaine, attribuant des poids différents aux couleurs primaires.
    $$ \text{Niveau de gris} = 0,299 \times R + 0,587 \times G + 0,114 \times B $$
*   **Moyenne Simple** : Méthode moins précise calculant la moyenne arithmétique des composantes.
    $$ \text{Niveau de gris} = (R + G + B) / 3 $$
*   **Max/Min** : Utilisation de la valeur maximale ou minimale entre les trois couleurs pour déterminer
tres de sortie.

### Formats et Précision

*   **8 bits** : Offre **256 nuances** de gris, allant de 0 (noir) à 255 (blanc). C'est le standard pour la plupart des images JPEG et PNG.
*   **16 bits** : Offre **65 535 étapes tonales**, utilisé pour le traitement d'images haute précision.
*   **32 bits** : Offre théoriquement plus de 4 milliards de nuances, souvent utilisé en virgule flottante pour les traitements internes dans des logiciels comme GIMP. gris

```python
gray_img = (
    red_arr_2D * 0.299
    + green_arr_2D * 0.587
    + blue_arr.reshape((1080, 1920)) * 0.114
)

gray_img = gray_img.astype(np.uint8)
plt.imshow(gray_img, cmap="gray")
```

## Exercice 3 — Simulation boursière vectorisée

#### Partie 1 — Génération

```python
variations = np.random.uniform(low=1-0.03, high=1+0.03, size=(1000,364))
variations = np.concatenate((np.ones((1000,1)), variations), axis=1)

actions = np.random.uniform(40, 260, (1000, 1))

print(variations.shape)
#print(variations)
#print(actions)
```

#### Partie 2 — Calculs

rendement moyen

Le **rendement moyen**, ou retour sur investissement moyen, représente le gain financier annuel réalisé sur un investissement pendant une période donnée, servant à évaluer la performance et à comparer le risque. Il existe deux méthodes principales de calcul selon le contexte :

*   **Taux de rendement moyen (TRM) ou comptable (ARR)** : Utilisé en comptabilité et budgétisation des immobilisations, il compare le bénéfice annuel moyen au coût initial. La formule est : $$\text{TRM} = \left( \frac{\text{Bénéfice annuel moyen}}{\text{Coût de l'investissement}} \right) \times 100\%$$. Cette méthode est simple mais ne tient pas compte de la valeur temporelle de l'argent ni de la volatilité.
*   **Rendement moyen annualisé (géométrique)** : Méthode privilégiée pour les investissements financiers (bourse, ETF) car elle intègre l'effet de composition. Pour un investissement unique sans flux de trésorerie, la formule est : $$\text{Rendement annualisé} = \left( \frac{\text{Valeur finale}}{\text{Valeur initiale}} \right)^{\frac{1}{\text{Années}}} - 1$$. Elle reflète la croissance réelle du capital au fil du temps, contrairement à la moyenne arithmétique qui a tendance à surestimer les résultats en cas de forte volatilité.

Les rendements historiques varient selon l'actif : les **actions** offrent en moyenne **7 à 10 %** nominal (soit **5 à 7 %** réel après inflation) sur le long terme, tandis que les **obligations** ou le **crowdfunding** doivent offrir un rendement supérieur au fonds euros pour compenser leur risque. Pour les portefeuilles avec des dépôts et retraits irréguliers, l'**IRR** (Internal Rate of Return) ou **XIRR** est la méthode la plus précise pour calculer le rendement annualisé pondéré en capital.

```python
variation_totale = variations.prod(axis=1)

#rendement moyen
avg_rendement = (variation_totale / variations[:, 0]) -1
#print(avg_rendement)

#pire action - ceci représente la variation de l'action initiale (soit 1)
worst_action = variation_totale.min()

#meilleure action
best_action = variation_totale.max()

#volatilité pour chaque action (std)
volatilité = variations.std(axis=1)
#print(volatilité)

print(worst_action)
print(best_action)
print('la meilleure action est à l\'indice:', np.where(variation_totale == best_action), '\n')
```
